**b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).**

```python
import pandas as pd
```

```python
file_path = '/content/train.csv'

# Data Loading using Pandas
df = pd.read_csv(file_path)

# Data Set Preview
df.head(11)
```

```python
#Identifying Missing Values in the Dataset
missing_values = df.isna()
missing_values_count = missing_values.sum()

# Displaying Missing Values Count
print("Missing values count in each column:")
print(missing_values_count)

#Checking for overall Missing Values
if missing_values.any().any():
    print("There are missing values in the DataFrame.")
else:
    print("No missing values in the DataFrame.")
```

**Removing the Units and Extracting the numerical values**

```python
df['Mileage']=df['Mileage'].str.split(' ').str[0]
df['Engine']=df['Engine'].str.split(' ').str[0]
df['Power']=df['Power'].str.split(' ').str[0]
df['Seats']=df['Seats'].fillna(0)
```

```python
df['Mileage'] = df['Mileage']. astype(float)
df['Engine'] = df['Engine']. astype(float)
df['Power'] = df['Power']. astype(float)
df['Seats'] = df['Seats'] . astype(float)
```

**Imputing the missing values with median**

```python
median_values = df[['Mileage', 'Engine', 'Power', 'Seats']].median()
df[['Mileage', 'Engine', 'Power', 'Seats']] = df[['Mileage', 'Engine', 'Power', 'Seats']].fillna(median_values)
```

**Since 'New_Price' has a significant number of missing values , we can drop it**

```python
df.drop(columns=['New_Price'], inplace=True)
```

```python
#Identifying Missing Values in the Dataset
missing_values = df.isna()
missing_values_count = missing_values.sum()

# Displaying Missing Values Count
print("Missing values count in each column:")
print(missing_values_count)

#Checking for overall Missing Values
if missing_values.any().any():
    print("There are missing values in the DataFrame.")
else:
    print("No missing values in the DataFrame.")
```

**C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value**

```python
import pandas as pd

encoded_fuel = pd.get_dummies(df['Fuel_Type'], prefix='Fuel_Type')

encoded_transmission = pd.get_dummies(df['Transmission'], prefix='Transmission')


data = pd.concat([df, encoded_fuel, encoded_transmission], axis=1)


data.drop(['Fuel_Type', 'Transmission'], axis=1, inplace=True)

print(data)
```
